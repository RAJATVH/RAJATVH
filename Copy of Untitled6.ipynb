```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("diabetes (2).csv")
df
```

```python
font1= {'family': 'serif', 'color': 'blue', 'size':20}
font2 = {'family': 'serif', 'color': 'darkred', 'size' :15}
plt.ylabel("Frequency", fontdict = font2)
plt.hist(df[ 'Glucose'], label="Glucose Level Histogram")
plt.xlabel("Glucose Range", fontdict = font2)
plt.legend(loc="best")
```

```python
newdf=df.groupby('Outcome')
non_diab=newdf.get_group(0)['SkinThickness'].mean()
diab=newdf.get_group(1)['SkinThickness'].mean()
plt.bar(['non_diab', 'diab'], [non_diab, diab], label="Revenue" )
plt.legend(loc="best")
```

```python
#2. Identify percentage of Diabetes and Non Diabetes patients in given dataset
newdf=df.groupby('Outcome')
non_diab=newdf.get_group(0)['Outcome'].count()
diab=newdf.get_group(1)['Outcome'].count()
tot=df['Outcome'].count()
print(non_diab,diab,tot)
array=[100*non_diab/tot,100*diab/tot]
plt.pie(array, labels=['Non diabetes','Diabetes'],autopct='%1.1ff%%')
plt.title('percentage of diabetes and non diabetesÂ patient')
```
