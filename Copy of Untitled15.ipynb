```python
def display_friends(friends):
    print("Name\t\tPhone Number")
    print("------------------------")
    for name, number in friends.items():
        print(f"{name}\t\t{number}")
    print()
```

```python
def add_friend(friends, name, number):
    friends[name] = number
    print("Friend added successfully!")
    print()
```

```python
def delete_friend(friends, name):
    if name in friends:
        del friends[name]
        print("Friend deleted successfully!")
    else:
        print("Friend not found in the dictionary.")
    print()
```

```python
def modify_number(friends, name, new_number):
    if name in friends:
        friends[name] = new_number
        print("Phone number modified successfully!")
    else:
        print("Friend not found in the dictionary.")
    print()
```

```python
def check_friend(friends, name):
    if name in friends:
        print(f"{name} is present in the dictionary.")
    else:
        print(f"{name} is not present in the dictionary.")
    print()
```

```python
def display_sorted_friends(friends):
    sorted_friends = sorted(friends.items())
    print("Name\t\tPhone Number")
    print("------------------------")
    for name, number in sorted_friends:
        print(f"{name}\t\t{number}")
    print()
```

```python
# Main program
friend_dict = {}
while True:
    print("1. Display all friends")
    print("2. Add a friend")
    print("3. Delete a friend")
    print("4. Modify phone number")
    print("5. Check if a friend is present")
    print("6. Display friends in sorted order")
    print("0. Exit")

    choice = input("Enter your choice: ")
    print()

    if choice == "1":
        display_friends(friend_dict)
    elif choice == "2":
        name = input("Enter friend's name: ")
        number = input("Enter friend's phone number: ")
        add_friend(friend_dict, name, number)
    elif choice == "3":
        name = input("Enter friend's name to delete: ")
        delete_friend(friend_dict, name)
    elif choice == "4":
        name = input("Enter friend's name to modify the phone number: ")
        new_number = input("Enter new phone number: ")
        modify_number(friend_dict, name, new_number)
    elif choice == "5":
        name = input("Enter friend's name to check: ")
        check_friend(friend_dict, name)
    elif choice == "6":
        display_sorted_friends(friend_dict)
    elif choice == "0":
        print("Exiting the Entry...")
        break
    else:
        print("Invalid choice! Please try again.")
    print()
```
